5. Calcula la correlación entre las variables de "fertility_rate" y "life_expectancy". Interpreta la relación entre estas dos variables. ¿Qué tipo de correlación hay? Explica la razón de esta relación

```python
sns.scatterplot(x = "fertility_rate", 
                y = "life_expectancy", 
                data = df, 
                size = 10,  
                color = "blue");
```

```python
correlacion = df[["fertility_rate", "life_expectancy"]].corr(method = "pearson")

correlacion.iloc[0,1]
```

Como se observa en el gráfico, hay una correlación lineal negativa entre la tasa de fertilidad y la esperanza de vida.

9. Genera un mapa de calor (heatmap) de la matriz de correlación de todas las variables numéricas y discute las relaciones más destacadas.

```python
df_correlaciones = df.corr(numeric_only=True, method = "spearman")

#Para visualziarlo en un mapa de calor:

# establecemos el tamaño de la gráfica
plt.figure(figsize = (15, 15))

mask = np.triu(np.ones_like(df_correlaciones, dtype=bool))

# convertimos nuestra matriz de correlación a una gráfica
sns.heatmap(df_correlaciones, # sobre que datos queremos hacer el heatmap
            annot = True, # para que nos muestre los valores de cada una de las celdas del heatmap
            fmt = ".2f",  # para que nos redondee a dos decimales
            cmap = "magma", # para establecer el color de la gráfica
            vmax = 1, # para indicar cual es el valor máximo de nuestra escala de colores
            vmin = -1,
            mask=mask); # para identificar el valor mínimo de nuestra escala de colores
```
